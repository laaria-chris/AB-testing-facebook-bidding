```python
# import libraries

from bs4 import BeautifulSoup
import requests
import time
import datetime

import smtplib
```

```python
url = "https://webscraper.io/test-sites/e-commerce/allinone/phones/touch"
```

```python
page = requests.get(url)
page
```

```python
soup = BeautifulSoup(page.text, "lxml")
soup
```

Title names

```python
product_names = soup.find_all("a", class_="title")
product_names
```

Prices

```python
Prices = soup.find_all("h4", class_="price float-end card-title pull-right")
Prices
```

Reviews

```python
import re


reviews = soup.find_all("p", class_="review-count float-end")
reviews
```

Description

```python
description = soup.find_all("p", class_="description card-text")
description
```

```python
product_name_list= []
for i in product_names:
    product_name_list.append(i.text.strip())
product_name_list
```

```python
price_list= []
for i in Prices:
    price_list.append(i.text.strip())
price_list
```

```python
reviews_list= []
for i in reviews:
    reviews_list.append(i.text.strip())
reviews_list
```

```python
description_list= []
for i in description:
    description_list.append(i.text.strip())
description_list
```

Put it into a dataframe- use pandas to access the dataframe.

```python
import pandas as pd
```

```python
table = pd.DataFrame({"Product Name": product_name_list, "Price": price_list, "Reviews": reviews_list, "Description": description_list})
table
```
